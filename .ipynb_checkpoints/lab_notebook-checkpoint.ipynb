```python
# EDA

import matplotlib.pyplot as plt

df['Price'].plot.box()
plt.show()
```

```python
plt.bar(df["No_of_Owners"].astype(str), df["Price"])
plt.title('Car Price by Number of Owners')
plt.xlabel('Number of Owners')
plt.ylabel('Car Price')
plt.show()
```
