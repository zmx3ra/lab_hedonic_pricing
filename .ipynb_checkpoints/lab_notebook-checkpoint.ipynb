```python
# Question 3

def maxmin(z):
    w = (z-np.min(z))/(np.max(z)-np.min(z))
    return w

y = df['Price'] # Create target variable vector
 # Create feature matrix
X = df.drop({'Price','Make','Color','Body_Type'}, axis=1) # Create feature matri
# Split the sample:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, # Feature and target variables
                                                    test_size=.2, # Split the sample 80 train/ 20 test
                                                    random_state=65) # For replication purposes
```

```python
# Question 4

x = df['Mileage_Run']
y = df['Price']

def slr(x,y): 
    x_bar = np.mean(x)
    y_bar = np.mean(y)
    b1 = np.inner(x-x_bar,y-y_bar)/np.inner(x-x_bar,x-x_bar)
    b0 = y_bar - b1*x_bar
    y_hat = b0 + b1*x
    residuals = y - y_hat
    return({'b0':b0,'b1':b1})

print(slr(x,y))

```

```python
# Question 6

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split


df['Year_Mileage'] = df['Make_Year'] * df['Mileage_Run']


y_log = np.log(df['Price'])


df['(Intercept)'] = 1


X = df.loc[:, ['(Intercept)', 'Mileage_Run', 'Make_Year', 'Year_Mileage']]
y = y_log


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=65
)


def mlr(X, y):
    XpX = X.T @ X            # X'X
    Xpy = X.T @ y            # X'y
    b = np.linalg.solve(XpX, Xpy)  # Solve normal equations
    y_hat = X @ b            # Predictions
    residuals = y - y_hat
    SSE = np.inner(residuals, residuals)
    rsq = 1 - SSE / np.inner(y - np.mean(y), y - np.mean(y))
    return {'b': b, 'y_hat': y_hat, 'residuals': residuals, 'rsq': rsq, 'SSE': SSE}


reg_train = mlr(X_train.to_numpy(), y_train.to_numpy())

print("MLR coefficients (train set):", reg_train['b'])
print("Train R^2:", reg_train['rsq'])


y_train_pred = X_train @ reg_train['b']
y_test_pred = X_test @ reg_train['b']

SSE = np.sum((y_test.to_numpy() - y_test_pred)**2)
SST = np.sum((y_test.to_numpy() - np.mean(y_test.to_numpy()))**2)
R2_test = 1 - SSE/SST

print("Test R^2:", R2_test)

train_rmse = np.sqrt(np.mean((y_train.to_numpy() - y_train_pred)**2))
test_rmse = np.sqrt(np.mean((y_test.to_numpy() - y_test_pred)**2))


print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)

# used AI to refine our mathematical functions and order of code cause we had one line in the wrong spot 
```

The R^2 train in this one is higher, and the RSME are going to be much different because this model does the log of price.

Part 7: We plotted a histogram for the price variable and noticed that it was right skewed. To address this in our new, more complex model, we transformed the variable using a log function. The resulting model is better than the previous models as indicated by a higher R^2 value. We cannot compare the RMSE, however, because the model was transformed. This shows that selecting the right model is important for machine learning because fitting a model to a dataset with skewed variables will produce less accurate results.

```python
# Question 8

from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=65
)

# alpha = regularization strength; increase alpha -> more shrinkage
lasso = Lasso(alpha=0.01)  # you can tune alpha
lasso.fit(X_train, y_train)


y_pred = lasso.predict(X_test)
 
# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

y_train_pred = lasso.predict(X_train)
y_test_pred = lasso.predict(X_test)

# RMSE
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

# R^2
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

print("Train RMSE:", train_rmse, " | Test RMSE:", test_rmse)
print("Train R^2:", train_r2, " | Test R^2:", test_r2)


#https://www.pythontutorials.net/blog/lasso-sklearn/#google_vignette
```

```python
coef = pd.Series(lasso.coef_, index=X.columns)
print("Lasso coefficients:")
print(coef)

# Identify best predictors (non-zero)
best_predictors = coef[coef != 0].sort_values(key=abs, ascending=False)
print("Best predictors (non-zero coefficients, sorted by impact):")
print(best_predictors)
```

```python
importance = lasso.coef_
feature_importance = pd.Series(importance, index=X.columns)  # Create series for readability
print("Feature Importance:\n", feature_importance)
```

The best feature is Make_Year (we presume because it is the closest to non-zero value.)
